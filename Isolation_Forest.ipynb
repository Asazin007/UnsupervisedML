```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
#Generate synthetic dtata
np.random.seed(42)
normal_data = np.random.normal(loc=0,scale=1,size=(100,2))
anomalous_data= np.random.normal(loc=4,scale=1,size=(10,2))
all_data = np.vstack((normal_data,anomalous_data))
```

```python
#Train the Isolation Forst
clf = IsolationForest(contamination=0.05, random_state=42)
clf.fit(all_data)
```

```python
#prediction of anamaly
predictions= clf.predict(all_data) #clear the current figure
anamaly_indices = np.where(predictions==-1)[0]
```

```python
#plot
plt.scatter(all_data[:,0],all_data[:,1],c='blue',label='Normal Data')
plt.scatter(all_data[anamaly_indices,0],all_data[anamaly_indices,1],c='red',label='Anamalous Data')
plt.legend()
plt.title('Anamaly Detection')
plt.ylabel('Feature 1')
plt.xlabel('Feature 2')
plt.show()

```
