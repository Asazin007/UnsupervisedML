```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons#dataset of moon shape 2D
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
X,_ =make_moons(n_samples=200, noise=0.1)
```

```python
#standarize
scaler= StandardScaler()
X_scaled= scaler.fit_transform(X)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show
```

```python
#Initialization the DBSCAN
Clus = DBSCAN(eps=0.4, min_samples=5)
labels = Clus.fit_predict(X_scaled)
```

```python
labels
```

```python
plt.scatter(X[:,0],X[:,1],c=labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Result for DBSCAN Clustering')
plt.show()
```
