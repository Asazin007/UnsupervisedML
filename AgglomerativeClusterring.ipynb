```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#synthetic data
data , labels = make_blobs(n_samples=300, centers=4, cluster_std=1.0,random_state=0)
#visualization
plt.scatter(data[:,0],data[:,1],c=labels,cmap='viridis')
plt.title("Original data")
plt.show()
```

```python
a1= AgglomerativeClustering(n_clusters=3)
labels = a1.fit_predict(data)

plt.scatter(data[:,0],data[:,1],c= labels, cmap='viridis')
plt.title('Agglomerative clustering')
plt.show()
```

```python
#create dendogram
linked = linkage(data,'ward')#ward is min variance criterion
dendrogram(linked)
plt.title("Clustering Dendrogram")
plt.show()
```
